```python
import os
import pandas as pd
import numpy as np

# 假设文件位于名为 "AutoY_Results_100" 的文件夹中，并且它们的名称遵循 "results_1.tsv", "results_2.tsv", ... 的模式
folder_path = '../AutoY_Results_100'  # 修改为实际的文件夹路径
num_files = 100  # 总共的文件数量

# 初始化空列表来存储每个文件的数据
all_data = []

# 遍历所有文件
for i in range(1, num_files + 1):
    file_path = os.path.join(folder_path, f'results_{i}.tsv')
    try:
        # 读取文件
        df = pd.read_csv(file_path, sep='\t', header=None)
        # 为每个疾病添加数据
        for index, row in df.iterrows():
            disease = row[0]
            metrics = row[1:].astype(float)  # 转换为浮点数
            all_data.append([disease] + metrics.tolist())
    except Exception as e:
        print(f"Error reading file {file_path}: {e}")

# 转换为 DataFrame
all_data_df = pd.DataFrame(all_data, columns=['Disease', 'Accuracy', 'Sensitivity', 'Specificity', 'AUC'])

# 计算每种疾病的平均值和标准差
results_summary = all_data_df.groupby('Disease').agg(['mean', 'std']).reset_index()

# 输出结果
results_summary

```

```python
import pandas as pd

# 读取上传的文件
file_path = '../AutoY_CSV_Results_100/CI51.csv'

# 加载数据
data = pd.read_csv(file_path)
data.head()
```

```python
# 根据用户的要求，重新绘制ROC曲线，保持颜色不变。
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# 疾病列表
diseases = data['Disease'].unique()
# 定义每种疾病对应的颜色，以保证颜色不会改变
disease_colors = {
    "RA": "blue",
    "T1D": "orange",
    "MS": "green",
    "IAA": "red"
}


# 设置画布
plt.figure(figsize=(10, 8))

# 为每种疾病绘制ROC曲线，保持颜色一致
for disease in diseases:
    disease_data = data[data['Disease'] == disease]
    fpr, tpr, _ = roc_curve(disease_data['Label'], disease_data['Pred'])
    roc_auc = auc(fpr, tpr)
    specificity = 1 - fpr
    plt.plot(specificity, tpr, label=f'{disease} (AUC = {roc_auc:.4f})', color=disease_colors[disease])

# 设置图例和坐标轴标签
plt.plot([0, 1], [1, 0], 'k--', color='grey')  # 修正了重复定义颜色的警告
plt.xlim([1.0, 0.0])  # 特异性从1降到0
plt.ylim([0.0, 1.05])
plt.xlabel('Specificity')
plt.ylabel('Sensitivity')
plt.legend(loc="lower right")

# 设置保存路径
save_path = '../R_roc/AutoY_roc_curve.png'  # 请根据需要修改路径

# 保存图表
plt.savefig(save_path, format='png')

# 显示图形
plt.show()

```
